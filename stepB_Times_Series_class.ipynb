```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
```

```python
# charger les données étiquetées pour chaque appareil ménager
label_train_dishwasher = pd.read_csv("StepTwo_LabelTrain_Dishwasher.csv")
label_train_kettle = pd.read_csv("StepTwo_LabelTrain_Kettle.csv")
label_train_microwave = pd.read_csv("StepTwo_LabelTrain_Microwave.csv")
label_train_tumbledryer = pd.read_csv("StepTwo_LabelTrain_TumbleDryer.csv")
label_train_washingmachine = pd.read_csv("StepTwo_LabelTrain_WashingMachine.csv")

# charger les données d'entrée
input_test_df = pd.read_csv("InputTest.csv")
input_train_df = pd.read_csv("InputTrain.csv")
```

```python
input_train_df = input_train_df.drop(["Index"], axis=1)
input_test_df = input_test_df.drop(["Index"], axis=1)

label_train_dishwasher = label_train_dishwasher.drop(["House_id","Index"], axis=1)
label_train_kettle = label_train_kettle.drop(["House_id","Index"], axis=1)
label_train_microwave = label_train_microwave.drop(["House_id","Index"], axis=1)
label_train_tumbledryer = label_train_tumbledryer.drop(["House_id","Index"], axis=1)
label_train_washingmachine = label_train_washingmachine.drop(["House_id","Index"], axis=1)
```

```python
def extract_features(data,time):
    features = pd.DataFrame()
    features["time"] = data.iloc[:,time]
    features['mean'] = data.mean(axis=1)
    features['median'] = data.median(axis=1)
    features['std'] = data.std(axis=1)
    features['House_id'] = data["House_id"]
    features['min'] = data.min(axis=1)
    features['max'] = data.max(axis=1)
    return features
```

```python
from imblearn.over_sampling import RandomOverSampler
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score

list_predictions_dishwasher = []
ros = RandomOverSampler(sampling_strategy='minority')

input_features = extract_features(input_train_df, 2)

x, y = ros.fit_resample(input_features, label_train_dishwasher["TimeStep_1"])

X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.3, random_state=42)


model_dishwasher = GradientBoostingClassifier(loss='deviance', learning_rate=0.1, n_estimators=100, subsample=1.0, criterion='friedman_mse', 
                               min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_depth=6, 
                               min_impurity_decrease=0.0, init=None, random_state=None, 
                               max_features=None, verbose=0, max_leaf_nodes=None, warm_start=False,
                               validation_fraction=0.1, n_iter_no_change=None, tol=0.0001, ccp_alpha=0.0)

model_dishwasher=model_dishwasher.fit(X_train, y_train)

preds = model_dishwasher.predict_proba(X_val)[:, 1]

# Calculer les métriques
precision = precision_score(y_val, preds.round())
recall = recall_score(y_val, preds.round())
f1 = f1_score(y_val, preds.round())

# Afficher les métriques
print(f'Precision : {precision:.4f}')
print(f'Recall : {recall:.4f}')
print(f'F1-score : {f1:.4f}')


for i in range(1,2161):
    # Extract features for the current timestep
    input_features = extract_features(input_test_df, i)
    print(i)
    # Make a prediction using the trained model
    predictions_dishwasher = model_dishwasher.predict_proba(input_features)[:,1]
    print((predictions_dishwasher.mean()))
    list_predictions_dishwasher.append(predictions_dishwasher)

df_predictions_dishwasher = pd.DataFrame(list_predictions_dishwasher)
df_predictions_dishwasher_T_flatter = df_predictions_dishwasher.T.values.flatten()
len(df_predictions_dishwasher_T_flatter)
```
